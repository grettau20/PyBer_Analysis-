```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan","Feb","Mar","Apl","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
fig = plt.figure()
ax = fig.add_subplot()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().

plt.plot(x_axis, y_axis, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0, 45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
plt.bar(x_axis, y_axis)
```

```python
#Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal plot
plt.barh(x_axis, y_axis)
```

```python
# Create the opposite axes plot
plt.barh(y_axis, x_axis)
```

```python
#inverting y axis in ascending order
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
# Add the legend.
plt.legend()
```
